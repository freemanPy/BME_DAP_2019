```python
# adatelemzos mantra
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Anomalia detekcio

```python
df=pd.read_csv("../data/Dataset_AnomalyDetection_MNW_40_100_publicdata.csv")
#ezt a path azt jelenti h 1 könyvtárral van kiljebb
df
```

```python
df.describe()
#szerencsénkre a max értékek közeliek oszlopok közt, így we can get away wo normalising
```

```python
from sklearn.cluster import KMeans
```

```python
model= KMeans()
```

```python
model
```

```python
model= KMeans(n_clusters=8, random_state=42)
```

```python
bemeno_valtozok= df.columns[:-2]
bemeno_valtozok
```

```python
model.fit(df[bemeno_valtozok])
```

```python
df['clusterID']=model.predict(df[bemeno_valtozok])
#ez egy szám array, melyik entry melyik clusterbe tartozik
df.head()
```

```python
#melyik clusternek mekkora a mérete
df.groupby("clusterID", as_index=False)['ID'].agg('count')
```

```python
#ne csak db számot, hanem hány ismert anomália van benne
df['info_2']=df['info'].apply(lambda x: 1 if x[0]=='d' else 0)
#if 'detected_anomaly'==1
 #   return 1
#else
 #   return 0
df
```

```python
df.groupby("clusterID", as_index=False).agg({'ID':'count', 'info_2':'mean'})
#df.groupby("clusterID", as_index=False).agg({'ID':['count', 'mean', 'max'], 'info2':'mean'})
#nekem nem jó az info_2 type ja, neki átlagot számolt rendesen
```

```python
for i in range(len(model.cluster_centers_)):
    plt.plot(list(model.cluster_centers_[i]), marker='o')
#melyik clusterben abban a dimenzióban mi a középpont model.cluster_centers_
```

```python
from sklearn.metrics.pairwise import euclidean_distances
dists= euclidean_distances(model.cluster_centers_)
dists
#ez nem az amit kerestünk
```

```python
uj_oszlopnevek=[]
for i in range(len(model.cluster_centers_)):
    df['cluster_'+str(i)]=0
    uj_oszlopnevek.append('cluster_'+str(i))

   
"""ez egy jó fv, jegyezd meg, pl 2D-s teret 3 számmal írsz le 3 távolság
3 ponttól, itt most 8 dimenzióra tömörít, 1. oszlop h a 0s clustertól mekkora távra van(?)"""
```
