## aggregacio

```python
aggreg_objektum=df.groupby(('visit_date', 'hetnapja'), as_index=False)
#ha ez true akkor csunya lesz a tablazat mert csak 1 oszlop lesz az adat, a tobbi multiindex lesz,
#ami most didaktikailag jobb mert nem kell elmagyaraznia a multiindexet pl
aggreg_objektum
```

```python
napistat=aggreg_objektum['visit_spend'].agg('sum')
napistat.columns=['visit_date', 'hetnapja','visit_spend_sum']
napistat.head()
```

```python
plt.plot(napistat['visit_spend_sum'])
```

```python
stat=napistat.groupby('hetnapja', as_index=False)['visit_spend_sum'].agg('mean')
```

```python
plt.plot(stat['hetnapja'], stat['visit_spend_sum'], marker='o')
```

```python
plt.scatter(stat['hetnapja'], stat['visit_spend_sum'], marker='o')
```

```python
plt.title('Cim')
plt.ylabel('felirat')
plt.hist(napistat['visit_spend_sum'], bins=20)
plt.savefig('dori.png')
#igy eddig a sorig kiadott abraparancsokbol renderel es kimenti kepfajlkent
plt.show()
```

```python
#pythonnak a joinja az a merge
eredmeny=napistat.merge(stat, on='hetnapja', how='inner')
eredmeny.head()
```

```python
eredmeny['error']=eredmeny['visit_spend_sum_x']-eredmeny['visit_spend_sum_y']
eredmeny.head()
```

```python
eredmeny['ape']=abs(eredmeny['error']/eredmeny['visit_spend_sum_y'])
```
